```python
def manual_lu_decomposition(matrix):
    """Perform LU decomposition of a square matrix manually without built-in functions."""
    n = len(matrix)  # Dimension of the square matrix
    L = [[0 for _ in range(n)] for _ in range(n)]  # Initialize L (lower triangular) with zeros
    U = [[0 for _ in range(n)] for _ in range(n)]  # Initialize U (upper triangular) with zeros

    # Step-by-step LU decomposition
    for i in range(n):
        # Calculate U (upper triangular matrix)
        for k in range(i, n):
            sum_lu = sum(L[i][j] * U[j][k] for j in range(i))  # Dot product for current row and column
            U[i][k] = matrix[i][k] - sum_lu

        # Calculate L (lower triangular matrix)
        for k in range(i, n):
            if i == k:
                L[i][i] = 1  # Diagonal elements of L are always 1
            else:
                sum_lu = sum(L[k][j] * U[j][i] for j in range(i))  # Dot product for current row and column
                L[k][i] = (matrix[k][i] - sum_lu) / U[i][i]

    return L, U

def calculate_determinant_from_lu(L, U):
    """Compute the determinant of a matrix using its LU decomposition."""
    # Determinant is the product of the diagonal elements of U
    det = 1
    for i in range(len(U)):
        det *= U[i][i]
    return det

# Example Input Matrix (replace with your matrix)
matrix = [
    [82, 56, 71, 26, 2, 27, 58, 22, 79, 1],
    [40, 0, 54, 93, 51, 6, 43, 73, 67, 52],
    [71, 0, 53, 88, 22, 94, 4, 28, 63, 73],
    [5, 0, 95, 7, 2, 39, 94, 27, 56, 47],
    [95, 0, 0, 11, 98, 59, 48, 87, 79, 10],
    [0, 80, 0, 36, 84, 18, 43, 10, 26, 63],
    [11, 0, 84, 0, 79, 32, 7, 29, 99, 56],
    [32, 0, 56, 0, 0, 53, 40, 50, 14, 47],
    [0, 0, 2, 0, 40, 0, 2, 83, 7, 5],
    [82, 56, 71, 26, 2, 27, 58, 22, 79, 1]
]

# Perform LU decomposition manually
L, U = manual_lu_decomposition(matrix)

# Print Results
print("Lower Triangular Matrix L:")
for row in L:
    print(row)

print("\nUpper Triangular Matrix U:")
for row in U:
    print(row)

# Calculate the Determinant
determinant = calculate_determinant_from_lu(L, U)
print(f"\nDeterminant of the matrix: {determinant}")
```

```python
#WIBF
def multiply(a, b):
    return a * b

def subtract(a, b):
    return a - b

def divide(a, b):
    return a / b

def lu_decomposition(matrix):
    n = len(matrix)
    # Initialize L and U matrices
    L = [[0 if i != j else 1 for j in range(n)] for i in range(n)]
    U = [[0] * n for _ in range(n)]

    for i in range(n):
        # Compute U matrix
        for j in range(i, n):
            sum_k = sum(multiply(L[i][k], U[k][j]) for k in range(i))
            U[i][j] = matrix[i][j] - sum_k

        # Compute L matrix
        for j in range(i + 1, n):
            sum_k = sum(multiply(L[j][k], U[k][i]) for k in range(i))
            L[j][i] = divide(subtract(matrix[j][i], sum_k), U[i][i])

    return L, U

def determinant_lu(matrix):
    # Perform LU decomposition
    _, U = lu_decomposition(matrix)

    # Calculate determinant as product of diagonal elements of U
    det = 1
    for i in range(len(U)):
        det *= U[i][i]

    return det

# Define the matrix correctly
matrix_determinant_example = [
    [54, 90, 86, 87, 26, 7, 49, 4, 5, 39],
    [67, 13, 42, 68, 15, 9, 41, 53, 72, 34],
    [0, 25, 34, 76, 24, 52, 76, 13, 55, 15],
    [0, 0, 76, 50, 33, 9, 1, 34, 42, 77],
    [0, 0, 0, 89, 49, 67, 3, 23, 78, 30],
    [0, 0, 0, 79, 9, 5, 93, 48, 27, 4],
    [0, 0, 0, 0, 37, 83, 95, 39, 79, 91],
    [0, 0, 0, 0, 0, 14, 64, 27, 4, 13],
    [0, 0, 0, 0, 0, 0, 92, 92, 20, 25],
    [0, 0, 0, 0, 0, 0, 0, 78, 64, 54]
]

print("Determinant using LU decomposition:", determinant_lu(matrix_determinant_example))
```
